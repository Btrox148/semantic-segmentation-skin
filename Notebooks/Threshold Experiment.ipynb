### Batch Thresholding

```python
image_fp = "../Data/Training_input/"

images = ['ISIC_0000000.jpg','ISIC_0000001.jpg','ISIC_0000003.jpg',
          'ISIC_0000004.jpg','ISIC_0000006.jpg','ISIC_0000007.jpg',
          'ISIC_0000008.jpg','ISIC_0000009.jpg','ISIC_0000011.jpg',
          'ISIC_0000012.jpg','ISIC_0000013.jpg','ISIC_0000014.jpg',
          'ISIC_0000015.jpg','ISIC_0000016.jpg','ISIC_0000017.jpg',]

tested = []

#for i in range(len(images)):
#    test_b = cv2.imread(image_fp + images[i], 1)
#    test_b = cv2.cvtColor(test_b, cv2.COLOR_BGR2RGB)
#    plt.figure('sampler', figsize=(10,10))
#    plt.subplot(5,3,i+1),plt.imshow(test_b)
#    plt.xticks([]),plt.yticks([])
#   plt.xlabel(images[i])
#   plt.savefig('../Plots/Images_input.png',bbox_inches='tight')
#   ret, thresh = cv2.threshold(test_b, 127, 255, cv2.THRESH_BINARY)
#   tested.append(thresh)

for i in range(len(images)):
    test_b = cv2.imread(image_fp + images[i], 1)
    test_b = cv2.cvtColor(test_b, cv2.COLOR_BGR2RGB)
    plt.figure('sampler', figsize=(1.5,1.5))
    plt.ylabel(""), plt.xlabel("")
    plt.imshow(test_b)
    plt.savefig('../Plots/THR/sample_'+str(i)+'.png', bbox_inches='tight')
    ret, thresh = cv2.threshold(test_b, 127, 255, cv2.THRESH_BINARY)
    tested.append(thresh)
```

```python
for i in range(len(images)):
    img = cv2.imread(image_fp + images[i],1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure('In_hist', figsize=(2,2))
    plt.xlabel('Intensidad de pixel')
    plt.ylabel('Recurrencia')
    plt.hist(img.ravel(), bins=255, fc='k', ec='k', density = True)
    plt.savefig('../Plots/THR/threshold_input_' + str(i) + '.png', bbox_inches='tight')
```

```python
for i in range(len(tested)):
    plt.figure('Out', figsize=(10,10))
    plt.subplot(5,3,i+1),plt.imshow(tested[i])
    plt.xticks([]),plt.yticks([])
    plt.xlabel(images[i])
    plt.savefig('../Plots/Images_output.png',bbox_inches='tight')
    
```

```python
for i in range(len(tested)):
    plt.figure('Out_hist', figsize=(10,10))
    plt.subplot(5,3,i+1),plt.hist(tested[i].ravel(), bins=8, fc='k', ec='k')
    plt.subplots_adjust(wspace=0.5, hspace=0.5)
    plt.xlabel('Intensidad de pixel')
    plt.ylabel('Recurrencia')
    plt.savefig('../Plots/threshold_output.png',bbox_inches='tight')
```
